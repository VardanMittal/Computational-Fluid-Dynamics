```python
import numpy as np
import matplotlib.pyplot as plt
import os, sys
sys.path.append(os.path.abspath(os.path.join('../..')))
```

```python
from utils.solvers import GaussElimination
```

```python
n = 100 # No of nodes
q= 1e6
k = 0.5
Delta_x = 0.02 / n
T1 = 100
T7 = 200
```

```python
b = np.zeros(n)
A = np.zeros((n,n))
```

```python
for i in range(1, n-1):
    b[i] = (-q * (Delta_x**2))/k
b[0] = - T1 + (-q * (Delta_x**2))/k
b[n-1] = - T7 + (-q * (Delta_x**2))/k
```

```python
for i in range(n-1):
    A[i][i] = -2
    A[i][i+1] = 1
    A[i+1][i] = 1

A[n-1][n-1] = -2
```

```python
x =GaussElimination(A,b)
```
